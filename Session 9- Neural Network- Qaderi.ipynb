## Neural Network Overview

A neural network is a powerful machine learning model inspired by the structure and functioning of the human brain. It consists of interconnected nodes, or "neurons," organized in layers. 

- **Input Layer**: This layer receives the initial data and passes it on to the subsequent layers.

- **Hidden Layers**: These intermediate layers process the input data through a series of weighted connections and apply activation functions to produce meaningful representations.

- **Output Layer**: This layer provides the final output or prediction based on the processed information from the previous layers.

Neural networks are capable of learning complex relationships within data and can be applied to a wide range of tasks, including classification, regression, image recognition, and natural language processing. During training, the network adjusts the weights on the connections to minimize the difference between predicted and actual outputs, a process known as backpropagation.

With deep learning, neural networks become even more powerful by incorporating many hidden layers, allowing them to learn intricate patterns in data. This makes them particularly effective for tasks with high complexity and large datasets.

<h2 style="color: #555; text-align: center;">Activation Functions in Neural Networks</h2>

<p>Activation functions are crucial components of neural networks. They introduce non-linearity to the model, enabling it to learn complex relationships in data. Here are some commonly used activation functions:</p>

<h3 style="color: #777;">1. <strong>Sigmoid Function</strong></h3>
<p>The sigmoid function squashes the input values between 0 and 1, making it suitable for binary classification tasks. However, it can suffer from the vanishing gradient problem.</p>

<h3 style="color: #777;">2. <strong>ReLU (Rectified Linear Unit)</strong></h3>
<p>ReLU sets all negative values to zero, allowing the network to learn faster and reducing the likelihood of the vanishing gradient problem. It's widely used in deep learning models.</p>

<h3 style="color: #777;">3. <strong>Tanh (Hyperbolic Tangent)</strong></h3>
<p>Tanh function squashes the input values between -1 and 1, which can be useful in certain scenarios. It addresses the vanishing gradient problem better than the sigmoid function.</p>

<h3 style="color: #777;">4. <strong>Softmax Function</strong></h3>
<p>Softmax is often used in the output layer for multi-class classification problems. It converts the raw output scores into probabilities, ensuring they sum up to 1.</p>

<h3 style="color: #777;">5. <strong>Leaky ReLU</strong></h3>
<p>Leaky ReLU allows a small, non-zero gradient for negative input values, addressing the 'dying ReLU' problem that can occur with standard ReLU.</p>

<h3 style="color: #777;">6. <strong>Swish Function</strong></h3>
<p>Swish is a recently proposed activation function that tends to perform well in deep neural networks, often outperforming ReLU.</p>

<p style="font-style: italic; text-align: center; color: #777;">Choosing the right activation function is a critical decision in designing neural networks, and it can significantly impact the model's performance.</p>

<h2 style="color: #555; text-align: center;">Loss Functions in Neural Networks</h2>

<p>Loss functions are crucial components in training neural networks. They quantify the error between predicted and actual values, guiding the model towards better performance. Here are some commonly used loss functions:</p>

<h3 style="color: #777;">1. <strong>Mean Squared Error (MSE)</strong></h3>
<p>MSE measures the average squared difference between predicted and actual values. It's widely used in regression tasks.</p>

<h3 style="color: #777;">2. <strong>Binary Cross Entropy</strong></h3>
<p>Binary Cross Entropy is used in binary classification tasks. It quantifies the difference between predicted probabilities and actual binary labels.</p>

<h3 style="color: #777;">3. <strong>Categorical Cross Entropy</strong></h3>
<p>Categorical Cross Entropy is suitable for multi-class classification problems. It quantifies the difference between predicted class probabilities and actual one-hot encoded labels.</p>

<h3 style="color: #777;">4. <strong>Mean Absolute Error (MAE)</strong></h3>
<p>MAE measures the average absolute difference between predicted and actual values. It's another commonly used loss function in regression.</p>

<h3 style="color: #777;">5. <strong>Huber Loss</strong></h3>
<p>Huber loss combines the characteristics of MSE and MAE, providing a balance between robustness to outliers and convergence speed.</p>

<p style="font-style: italic; text-align: center; color: #777;">Selecting the appropriate loss function depends on the nature of the problem and the type of output the neural network is expected to produce.</p>

```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
from tensorflow.keras import layers, models


# Step 1: Load and Explore the MNIST Dataset

# Load the MNIST dataset, which consists of 28x28 grayscale images of handwritten digits (0-9).
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
# The training dataset contains 60,000 images, and the test dataset contains 10,000 images.
print("Number of training samples:", len(train_images))
print("Number of test samples:", len(test_images))
```

```python
# Show an example image from the dataset
plt.figure()
plt.imshow(train_images[0], cmap='gray')
plt.title(f"Label: {train_labels[0]}")
plt.show()
```

```python
# Step 2: Preprocess the Data

# Reshape the images to have a single color channel (grayscale) and normalize pixel values to [0, 1].
train_images = train_images.reshape((60000, 28, 28, 1))
test_images = test_images.reshape((10000, 28, 28, 1))
train_images = train_images.astype('float32') / 255
test_images = test_images.astype('float32') / 255

# One-hot encode the labels.
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

```python
# Select a few random images from the training dataset
sample_images = train_images[:5]

# Display the selected images along with their labels
plt.figure(figsize=(12, 5))

for i, image in enumerate(sample_images):
    plt.subplot(1, 5, i + 1)
    plt.imshow(image.reshape(28, 28), cmap='gray')
    plt.title(f"Label: {train_labels[i].argmax()}")

plt.show()
```

```python
# Step 3: Build the Deep Learning Model

# Create a sequential model, which allows us to build a neural network layer by layer.
model = models.Sequential()

# Add the first convolutional layer with 32 filters, each of size (3, 3), and ReLU activation.
# Input shape is (28, 28, 1) for grayscale images.
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))

# Add a max-pooling layer to downsample the spatial dimensions.
model.add(layers.MaxPooling2D((2, 2)))

# Add another convolutional layer with 64 filters and ReLU activation.
model.add(layers.Conv2D(64, (3, 3), activation='relu'))

# Add another max-pooling layer.
model.add(layers.MaxPooling2D((2, 2)))

# Add a third convolutional layer with 64 filters and ReLU activation.
model.add(layers.Conv2D(64, (3, 3), activation='relu'))

# Flatten the output from 2D to 1D for the fully connected layers.
model.add(layers.Flatten())

# Add a dense (fully connected) layer with 64 units and ReLU activation.
model.add(layers.Dense(64, activation='relu'))

# Add the output layer with 10 units (one for each digit 0-9) and softmax activation.
model.add(layers.Dense(10, activation='softmax'))
```

```python
# Step 4: Compile the Model

# Compile the model with the Adam optimizer, categorical cross-entropy loss for multi-class classification,
# and metrics to measure during training (accuracy in this case).
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Step 5: Train the Model

# Train the model on the training data with 5 epochs and a batch size of 64.
history = model.fit(train_images, train_labels, epochs=5, batch_size=64,
                    validation_data=(test_images, test_labels))
```

```python
plt.plot(history.history['accuracy'], label='Acuuracy')

plt.plot(history.history['val_accuracy'], label='Validation Accuracy')

plt.legend()

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Accuracy Over Epochs')

plt.show()
```
