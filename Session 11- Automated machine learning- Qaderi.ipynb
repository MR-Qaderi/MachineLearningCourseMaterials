## Regression

```python
from pycaret.regression import *
import pandas as pd
```

```python
from pycaret.datasets import get_data
boston = get_data("boston")
```

### Model Setup

```python
setup(data = boston , target = 'medv', session_id = 123)
```

```python
compare_models()
```

```python
ada = create_model("ada")
```

```python
plot_model(ada)
```

```python
evaluate_model(ada)
```

```python
dt = create_model("dt")
tuned_dt = tune_model(dt)
```

```python
evaluate_model(tuned_dt)
```

```python
best = automl()
```

```python
best
```

## Classification

```python
import pandas as pd
from pycaret.classification import *
```

```python
juice = get_data("juice")
```

Model Setup

```python
setup(data = juice, target = "Purchase")
```

```python
compare_models()
```

```python
rf = create_model("rf")
```

```python
plot_model(estimator= rf, plot = "auc")
```

```python
plot_model(estimator = rf, plot = "feature")
```

## Ensemble

```python
knn = create_model("knn")
lr = create_model("lr")

blend = blend_models(estimator_list = [rf, knn, lr])
```

```python
plot_model(estimator= blend, plot = "auc")
```

### Evaluation

```python
plot_model(estimator= knn, plot = "confusion_matrix")
```

```python
pred = predict_model(rf)
```

```python
evaluate_model(rf)
```

## Clustering
