## Handling Outliers in Data

### What are Outliers?
Outliers are data points that significantly deviate from the rest of the data in a dataset. They can be unusually high or low values that may skew the analysis or modeling results.

### Importance of Handling Outliers
Handling outliers is important for several reasons:
- Outliers can adversely affect the performance and accuracy of machine learning models.
- They can lead to incorrect statistical analysis and misleading insights.
- Outliers can impact the distribution and assumptions of the data, affecting the validity of results.

### Ways to Handle Outliers
There are various methods to handle outliers:
1. **Z-Score Method**: Calculate the Z-Score for each data point and remove those with Z-Scores above a certain threshold.
2. **Transformation**: Apply mathematical transformations like logarithmic or square root transformations to reduce the impact of outliers.
3. **Capping**: Set a predefined cap or floor value for data points beyond a certain range.
4. **Removing**: Simply remove the outliers from the dataset.
5. **Binning**: Group outliers into a separate category or bin.

### Z-Score vs. Transformation
- **Z-Score** identifies outliers based on their deviation from the mean. It does not alter the actual values but flags outliers for potential removal.
- **Transformation** changes the scale or distribution of data values. It aims to reduce the impact of outliers and improve data characteristics.

The choice of method depends on the nature of the data and the goals of analysis or modeling.


```python
house = pd.read_csv("/content/House Prices.csv")
house
```

### Z-SCORE METHOD

<img src="https://www.gigacalculator.com/img/calculators/z-score-distribution-with-quantiles.png" />

```python
# Calculate Z-Scores for a specific colum
z_scores = np.abs((house["LotArea"] - np.mean(house["LotArea"])) / np.std(house["LotArea"]))

# Define the Z-Score threshold for outlier detection
threshold = 3

# Create a new DataFrame with outliers removed based on Z-Score
house_no_outliers_z = house[z_scores < threshold]
house_no_outliers_z
```

### IQR Method

<img src="https://th.bing.com/th/id/R.aadee74e39041b5281e392ea11782072?rik=JyqEuOtbG0dJhA&pid=ImgRaw&r=0" />

```python
# Calculate quartiles
Q1 = np.percentile(house["LotArea"], 25)
Q3 = np.percentile(house["LotArea"], 75)

# Calculate IQR (Interquartile Range)
IQR = Q3 - Q1

# Define the outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Create a new array with outliers removed based on IQR
house_no_outliers_iqr = house[(house["LotArea"] >= lower_bound) & (house["LotArea"] <= upper_bound)]
house_no_outliers_iqr
```

## Importance of Scaling in Machine Learning

Scaling is a crucial preprocessing step in machine learning that aims to bring all the features or variables of your dataset to a similar scale. This is essential because many machine learning algorithms are sensitive to the scale of input features, leading to various issues:

- **Algorithm Performance:** Some algorithms, such as gradient descent-based optimization, converge faster when features are on a similar scale. Dominant features can hinder the learning process.

- **Distance-Based Algorithms:** Algorithms like k-nearest neighbors (KNN) and clustering rely on distance metrics. Features with different scales can disproportionately influence these algorithms.

- **Regularization:** Methods like L1 and L2 regularization can be affected by feature scale. Features with larger scales might have a more substantial impact on the regularization term.

### Common Scaling Methods

1. **Min-Max Scaling (Normalization):** Scales features to a specific range (e.g., 0 to 1). Formula: `(X - X_min) / (X_max - X_min)`.

2. **Standardization (Z-Score Scaling):** Transforms features to have zero mean and unit variance. Formula: `(X - mean) / std_dev`.

3. **Robust Scaling:** Similar to standardization, but uses the median and interquartile range (IQR) for scale. More robust to outliers.

4. **Log Transformation:** Compresses a wide range of values into a smaller scale using logarithms. Useful for skewed data distributions.

### Scaling with Scikit-Learn

Scikit-Learn provides convenient tools for scaling:

- `StandardScaler`: Standardization using mean and standard deviation.
- `MinMaxScaler`: Min-Max scaling to a specified range.
- `RobustScaler`: Robust scaling using median and IQR.
- `PowerTransformer`: Applies power transformations like the log transform.

Remember to choose the appropriate scaling method based on your data and the requirements of your machine learning algorithm. Experimentation and evaluation are key to finding the best scaling approach for your model's performance.


```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Create the sample dataset
np.random.seed(42)
data = {
    'ID': np.arange(1, 101),
    'Name': ['John', 'Alice', 'Bob', 'Mary', 'David'] * 20,
    'Age': np.random.randint(18, 65, size=100),
    'Gender': ['Male', 'Female'] * 50,
    'Income': np.random.randint(20000, 100000, size=100),
    'City': ['New York', 'Los Angeles', 'Chicago', 'Miami', 'Houston'] * 20,
    'Education': ['High School', 'Bachelor', 'Master', 'PhD', 'Bachelor'] * 20,
    'Marital_Status': ['Married', 'Single', 'Married', 'Single', 'Divorced'] * 20,
    'Credit_Score': np.random.randint(300, 850, size=100),
    'Loan_Amount': np.random.randint(5000, 50000, size=100),
    'Default': np.random.choice([0, 1], size=100)
}

df = pd.DataFrame(data)

# Select columns for scaling
columns_to_scale = ['Age', 'Income', 'Credit_Score', 'Loan_Amount']

# Create separate DataFrames for scaled data
df_minmax_scaled = df.copy()
df_standard_scaled = df.copy()

# Apply Min-Max Scaling to 'Age' and 'Income'
minmax_scaler = MinMaxScaler()
df_minmax_scaled[['Age', 'Income']] = minmax_scaler.fit_transform(df_minmax_scaled[['Age', 'Income']])

# Apply Standard Scaling to 'Credit_Score' and 'Loan_Amount'
standard_scaler = StandardScaler()
df_standard_scaled[['Credit_Score', 'Loan_Amount']] = standard_scaler.fit_transform(df_standard_scaled[['Credit_Score', 'Loan_Amount']])
```

```python
# Display the Standard scaled data
print("\nStandard Scaled Data:")
df_standard_scaled
```

# encoding-categorical-data

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder, LabelBinarizer

# Create the sample dataset
data = {
    'Gender': ['Male', 'Female', 'Male', 'Male', 'Female'],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Miami', 'Houston'],
    'Education': ['High School', 'Bachelor', 'Master', 'PhD', 'Bachelor'],
    'Marital_Status': ['Married', 'Single', 'Married', 'Single', 'Divorced'],
    'Payment_Method': ['Credit Card', 'PayPal', 'Credit Card', 'Bitcoin', 'PayPal'],
    'Browser': ['Chrome', 'Firefox', 'Safari', 'Chrome', 'Edge'],
    'OS': ['Windows', 'MacOS', 'Linux', 'Windows', 'MacOS'],
    'Membership_Level': ['Silver', 'Gold', 'Bronze', 'Gold', 'Silver'],
    'Newsletter_Subscribed': ['Yes', 'No', 'Yes', 'No', 'Yes']
}

df = pd.DataFrame(data)
df.head()
```

```python
label_encoder = LabelEncoder()
df['Gender_LabelEncoded'] = label_encoder.fit_transform(df['Gender'])
df.head()
```

```python
one_hot_encoder = OneHotEncoder(sparse=False)
one_hot_encoded = one_hot_encoder.fit_transform(df[['City']])
df = pd.concat([df, pd.DataFrame(one_hot_encoded, columns=one_hot_encoder.get_feature_names_out(['City']))], axis=1)
df.head()
```

```python
ordinal_encoder = OrdinalEncoder(categories=[['High School', 'Bachelor', 'Master', 'PhD']])
df['Education_OrdinalEncoded'] = ordinal_encoder.fit_transform(df[['Education']])
df.head()
```

```python
df = df.drop(['Gender', 'City', 'Education'], axis=1)
df.head()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load and prepare the Iris dataset
iris = load_iris()
x = iris.data  # Features
y = iris.target  # Target (species)
feature_names = iris.feature_names

print("Original Iris Data:")
pd.DataFrame(x, columns=feature_names).head()
```

```python
# Standardize the features to have zero mean and unit variance
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

print("\nStandardized Iris Data:")
pd.DataFrame(x_scaled, columns=feature_names).head()
```

```python
# Perform PCA
pca = PCA(n_components = 2)
x_pca = pca.fit_transform(x_scaled)

print("\nPCA Transformed Data:")
df_pca = pd.DataFrame(data = x_pca, columns = ['principal component 1', 'principal component 2'])
df_pca
```

```python
iris_pca = pd.concat([df_pca, df[['Species']]], axis = 1)
iris_pca
```

```python
# Visualize the PCA components
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1)
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = iris_pca['Species'] == target
    ax.scatter(iris_pca.loc[indicesToKeep, 'principal component 1']
               , iris_pca.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```
